```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
url = "https://www.cricbuzz.com/cricket-stats/icc-rankings/men/teams"
```

```python
page = urlopen(url)
```

```python
page
```

```python
html_bytes = page.read()
html = html_bytes.decode('utf-8')
#print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
#print(soup)
```

```python
prettyHTML = soup.prettify()
#print(prettyHTML)
```

```python
position = []
for i in soup.findAll("div",{"class":"cb-col cb-col-20 cb-lst-itm-sm"}):
  position.append(i.text)
print(position)
```

```python
team = []
for i in soup.findAll("div",{"class":"cb-col cb-col-50 cb-lst-itm-sm text-left"}):
  team.append(i.text)
print(team)
```

```python
rating = []
points = []

for idx,tags in enumerate(soup.findAll("div",{"class":"cb-col cb-col-14 cb-lst-itm-sm"})):
  if idx % 2 == 0:
    rating.append(tags.text)
  else:
    points.append(tags.text)
print(rating)
print(points)
```

```python
df = pd.DataFrame(
    {
        'Position':position,
        'Team' : team,
        'Ratings':rating,
        'Points': points
    }
)
```
